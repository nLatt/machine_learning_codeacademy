```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
```

```python
# load the handwritten digits
train_digits = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/optdigits/optdigits.tra", header=None)
test_digits = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/optdigits/optdigits.tes", header=None)

digits = datasets.load_digits()
```

```python
labels = test_digits[64]
# print(list(labels))

test_digits = test_digits.drop(columns=[64])
train_digits = train_digits.drop(columns=[64])
```

```python
fig = plt.figure(figsize=(6, 6))
for i in range(64):
    ax = fig.add_subplot(8, 8, i+1, xticks=[], yticks=[])
    ax.imshow(digits.images[i], cmap=plt.cm.binary)
    ax.text(0, 7, str(digits.target[i]))
plt.show()
```

```python
inertias = []
k_count = [i for i in range(1, 30+1)]

for k in k_count:
    model = KMeans(n_clusters=k)
    model.fit(train_digits)
    inertias.append(model.inertia_)
```

```python
plt.plot(k_count, inertias)
plt.show()
```

```python
best_k = 10
```

```python
model = KMeans(n_clusters=best_k)
model.fit(train_digits)
results = model.predict(test_digits)
```

```python
number = []
unique_elements, counts_element = np.unique(results, return_counts=True)
counts_element
```

```python
df = pd.DataFrame({"labels": labels, "groups": results})
ct = pd.crosstab(df["groups"], df["labels"])
print(ct)
```

```python
fig = plt.figure(figsize=(8, 3))
fig.suptitle("cluster centers")

for i in range(10):
    ax = fig.add_subplot(2, 5, 1 + i)
    ax.imshow(model.cluster_centers_[i].reshape((8, 8)), cmap=plt.cm.binary)
```
