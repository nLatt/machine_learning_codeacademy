```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
from pandas.plotting import scatter_matrix
```

```python
data = pd.read_csv("tennis_stats.csv")
data.head()
```

```python
# explore data with hist, heatmap, corr and scatter_matrix

data.hist(bins=50, figsize=(24, 24))
plt.show()
```

```python
plt.figure(figsize=(20,5))

win_win = plt.subplot(1, 2, 1)
plt.scatter(data["Winnings"], data["Wins"])
plt.xlabel("Winnings")
plt.ylabel("Wins")

plt.subplot(1, 2, 2, sharex=win_win)
plt.scatter(data["Winnings"], data["Losses"])


# plt.ylim(0, 1)
plt.show()
```

```python
data_corr = data.corr()
data_corr["Ranking"].sort_values(ascending=False)
```

```python
fig, ax = plt.subplots(figsize=(15, 15))
sns.heatmap(data_corr, ax=ax, cmap="plasma")
plt.show()
```

```python
features = ["Winnings",
            "DoubleFaults",
            "BreakPointsOpportunities",
            "BreakPointsFaced",
            "Aces",
            "Wins"]

scatter_matrix(data[features], figsize=(12, 12))
plt.show()
```
