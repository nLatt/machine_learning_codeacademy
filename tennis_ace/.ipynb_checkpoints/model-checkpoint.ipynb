```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as nd
import matplotlib.pyplot as plt
```

```python
# reading CSV
data = pd.read_csv("tennis_stats.csv")
data.head()
```

```python
# Extracting essential data with highest negative correlation because a lower rank = better
essential_columns = ["TotalPointsWon",
                     "Aces",
                     "Wins",
                     "Losses",
                     "DoubleFaults",
                     "BreakPointsFaced",
                     "ReturnGamesPlayed",
                     "ServiceGamesPlayed",
                     "Winnings",
                     "BreakPointsOpportunities"]
outcome_data = data["Ranking"]
essential_data = data[essential_columns]
```

```python
# Select specific features for the first LR model

features = essential_data[["BreakPointsOpportunities"]]#.values.reshape(-1,1)

print(type(essential_data["TotalPointsWon"]))
```

```python
# Spliting the data

features_train, features_test, outcome_train, outcome_test = train_test_split(features, outcome_data, train_size=0.8)
```

```python
# Creating and training a model

model = LinearRegression()
model.fit(features_train, outcome_train)
model.score(features_test, outcome_test)
```

```python
# Iterating through every essential feature

def iterator(features_train, features_test, outcome_train, outcome_test):
    model = LinearRegression()
    model.fit(features_train, outcome_train)
    score = model.score(features_test, outcome_test)
    return score

score_list = []
    
for element in essential_columns:
    features = essential_data[[element]]
    features_train, features_test, outcome_train, outcome_test = train_test_split(features, outcome_data, train_size=0.8)
    score_list.append(iterator(features_train, features_test, outcome_train, outcome_test))
    
best_feature = score_list.index(max(score_list))
print(essential_columns[best_feature])
```

```python
pred = model.predict(features_test)
plt.scatter(outcome_test, pred)
plt.show()
```

```python
plt.plot(score_list, range(1, len(score_list)+1))
plt.show()
```
